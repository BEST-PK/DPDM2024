# **HW 2 หา pattern ที่น่าสนใจในหนังอินเดีย**

สนใจ Director, Genre ในการนำมาหา pattern ที่น่าสนใจในหนังอินเดีย
"ผู้กำกับสร้างหนังแนวไหนออกมาฉายคู่กันบ้าง ?"

```python
movies_india = pd.read_csv('/content/drive/MyDrive/dpdm24_DATA/IMDb Movies India.csv' , encoding='latin-1')
movies_india.head()
```

```python
movies_india = movies_india[['Director','Genre']]
movies_india
```

**การตรวจสอบค่าว่าง isnull()**

```python
movies_india[['Director', 'Genre']].isnull().any() # Check for nulls in the specified columns.
```

จัดการค่าว่าง

โดยลบออก เฉพาะ column ที่เราสนใจคือ Director, Genre ในการนำมาหา pattern

```python
movies_india = movies_india.dropna(subset=['Director','Genre'])
movies_india
```

```python
movies_india.groupby('Director')['Genre'].apply(set).apply(list)
```

```python
#transactions1= movies_india.groupby('Year')['Genre'].apply(set).apply(list)  #.reset_index()
#transactions1
transactions_india_movies= movies_india.groupby('Director')['Genre'].apply(set).apply(list)  #.reset_index()
transactions_india_movies
```

**วิเคราะห์ Parterns**

```python
!pip install apyori
import pandas as pd
from apyori import apriori
```

```python
frequentItemsets = list(apriori(transactions_india_movies,min_support = 0.03))
frequentItemsets
```

```python
len(frequentItemsets)
```

```python
min_sup = 0.03

frequentItemsets_2 = list(apriori(transactions_india_movies,min_support = min_sup))
for i in frequentItemsets_2:
    if len(i[0])>1:
        print('yes')
        print(i)
```
